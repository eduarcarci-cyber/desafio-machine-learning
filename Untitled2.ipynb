```python
# ============================================================
# SISTEMA DE RECOMENDAÇÃO DE PRODUTOS
# DESAFIO N1 - MACHINE LEARNING
# ============================================================

# ============================================================
# 1. IMPORTAÇÃO DAS BIBLIOTECAS
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

# ============================================================
# 2. CRIAÇÃO DO DATASET FICTÍCIO
# ============================================================

np.random.seed(42)

usuarios = range(1, 21)
produtos = range(1, 16)

dados = []

for usuario in usuarios:
    for produto in produtos:
        if np.random.rand() > 0.6:
            dados.append([
                usuario,
                produto,
                np.random.randint(1, 6)
            ])

df = pd.DataFrame(
    dados,
    columns=['user_id', 'product_id', 'rating']
)

# ============================================================
# 3. VISUALIZAÇÃO DOS DADOS
# ============================================================

print("Primeiras linhas do dataset:")
print(df.head())

print("\nInformações gerais:")
print(df.info())

print("\nValores nulos:")
print(df.isnull().sum())

# ============================================================
# 4. NORMALIZAÇÃO DOS DADOS
# ============================================================

scaler = MinMaxScaler()

df[['rating']] = scaler.fit_transform(
    df[['rating']]
)

print("\nDados normalizados:")
print(df.head())

# ============================================================
# 5. MATRIZ USUÁRIO x PRODUTO
# ============================================================

matriz = df.pivot_table(
    index='user_id',
    columns='product_id',
    values='rating'
).fillna(0)

print("\nMatriz usuário x produto:")
print(matriz.head())

# ============================================================
# 6. MODELO DE RECOMENDAÇÃO KNN
# ============================================================

modelo_knn = NearestNeighbors(
    metric='cosine',
    algorithm='brute'
)

modelo_knn.fit(matriz)

# ============================================================
# 7. VALIDAÇÃO CRUZADA
# ============================================================

X = matriz.values

# Criando alvo fictício
y = np.random.randint(0, 2, size=X.shape[0])

modelo_cv = LogisticRegression(
    max_iter=1000
)

scores = cross_val_score(
    modelo_cv,
    X,
    y,
    cv=5
)

print("\nResultado da validação cruzada:")
print(scores)

print("\nMédia da validação cruzada:")
print(scores.mean())

# ============================================================
# 8. TREINAMENTO E TESTE
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

modelo_cv.fit(X_train, y_train)

y_pred = modelo_cv.predict(X_test)

# ============================================================
# 9. MÉTRICAS DE AVALIAÇÃO
# ============================================================

print("\nMÉTRICAS DO MODELO")

print("Accuracy:")
print(accuracy_score(y_test, y_pred))

print("\nPrecision:")
print(precision_score(y_test, y_pred))

print("\nRecall:")
print(recall_score(y_test, y_pred))

print("\nF1-Score:")
print(f1_score(y_test, y_pred))

# ============================================================
# 10. MATRIZ DE CONFUSÃO
# ============================================================

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues'
)

plt.title("Matriz de Confusão")
plt.xlabel("Previsto")
plt.ylabel("Real")

plt.show()

# ============================================================
# 11. RECOMENDAÇÃO DE PRODUTOS
# ============================================================

distancias, indices = modelo_knn.kneighbors(
    [matriz.iloc[0]],
    n_neighbors=6
)

print("\nUsuários semelhantes ao usuário 1:")
print(indices)

# ============================================================
# 12. SALVAR DATASET CSV
# ============================================================

df.to_csv(
    "interacoes_usuarios.csv",
    index=False
)

print("\nDataset salvo com sucesso!")

# ============================================================
# 13. CONCLUSÃO
# ============================================================

print("""
Projeto finalizado com sucesso!

Foi desenvolvido um sistema de recomendação
utilizando Machine Learning com o algoritmo KNN.

O projeto realizou:
- Criação de dataset fictício
- Limpeza de dados
- Normalização
- Treinamento do modelo
- Avaliação de métricas
- Visualização de resultados
""")
```
